```python
from pathlib import Path
import pandas as pd
```

```python
root = Path("D:/workspace/image-to-image/results/sdo")
```

```python
class fake_dir:
    def __init__(self, name, path):
        self.name = name
        self.path = path

version = "version_0"
ckpt = "last"
timesteps = 10

fg = lambda x: sorted(Path(x).glob(f"**/{ckpt}*/csv"))[0]
fd = lambda x: sorted(Path(x).glob(f"**/{ckpt}*_timesteps={timesteps}/csv"))[0]

fake_roots = [
    fake_dir("pix2pix",                Path(fg(root/f"gan/pix2pix/logs/{version}/eval/"))),
    fake_dir("pix2pixHD",              Path(fg(root/f"gan/pix2pixhd/logs/{version}/eval/"))),
    fake_dir("pix2pixCC",              Path(fg(root/f"gan/pix2pixcc/logs/{version}/eval/"))),
    fake_dir("DDPM (x0)",           Path(fd(root/f"diffusion/ddpm/x0/logs/{version}/eval/"))),
    fake_dir("Fast-DDPM (x0)",      Path(fd(root/f"diffusion/fast_ddpm/x0/logs/{version}/eval/"))),
    fake_dir("DDPM (eps)",      Path(fd(root/f"diffusion/ddpm/noise/logs/{version}/eval/"))),
    fake_dir("Fast-DDPM (eps)", Path(fd(root/f"diffusion/fast_ddpm/noise/logs/{version}/eval/"))),
]

fake_roots_ema = [
    fake_dir("pix2pix EMA",                Path(fg(root/f"gan/pix2pix/logs/{version}/eval_ema/"))),
    fake_dir("pix2pixHD EMA",              Path(fg(root/f"gan/pix2pixhd/logs/{version}/eval_ema/"))),
    fake_dir("pix2pixCC EMA",              Path(fg(root/f"gan/pix2pixcc/logs/{version}/eval_ema/"))),
    fake_dir("DDPM (x0) EMA",           Path(fd(root/f"diffusion/ddpm/x0/logs/{version}/eval_ema/"))),
    fake_dir("Fast-DDPM (x0) EMA",      Path(fd(root/f"diffusion/fast_ddpm/x0/logs/{version}/eval_ema/"))),
    fake_dir("DDPM (eps) EMA",      Path(fd(root/f"diffusion/ddpm/noise/logs/{version}/eval_ema/"))),
    fake_dir("Fast-DDPM (eps) EMA", Path(fd(root/f"diffusion/fast_ddpm/noise/logs/{version}/eval_ema/"))),
]
```

```python
class Metric:
    def __init__(self, fake_roots, stage="test"):
        self.fake_roots = fake_roots
        self.stage = stage

    def get_summary_metrics(self):
        for fake_root in self.fake_roots:
            df = pd.read_csv(fake_root.path/self.stage/f"{self.stage}_metrics.csv")
            df["Name"] = fake_root.name
            cols = df.columns.tolist()
            cols = cols[-1:] + cols[:-1]
            df = df[cols]
            yield df

    def get_metrics(self, name="metrics_normalized_all"):
        for fake_root in self.fake_roots:
            yield pd.read_csv(fake_root.path/self.stage/f"{name}.csv")
```

```python
m = Metric(fake_roots)
df_sum = pd.concat(m.get_summary_metrics())
df_sum
```

```python
m_ema = Metric(fake_roots_ema)
df_sum_ema = pd.concat(m_ema.get_summary_metrics())
df_sum_ema
```

```python
def get_table(df, type_name, metric_name, ascending=False):
    print(f"{type_name}")
    df = df[df["Type"] == type_name].sort_values(metric_name, ascending=ascending)
    df = df.drop(columns=["Type"])
    return df
```

```python
type_name = "Normalized_All"
metric_name = "Pearson CC"
```

```python
get_table(df_sum, type_name, metric_name, ascending=False)
```

```python
get_table(df_sum_ema, type_name, metric_name, ascending=False)
```

```python
get_table(pd.concat([df_sum, df_sum_ema]), type_name, metric_name, ascending=False)
```
