```python
from sunpy.map import Map
import astropy.units as u
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
date = '2020-01-01T00:00:00'
```

```python
sdo_euv_94 = Map(f'sdo/94/{date}.fits')
sdo_euv_131 = Map(f'sdo/131/{date}.fits')
sdo_euv_171 = Map(f'sdo/171/{date}.fits')
sdo_euv_193 = Map(f'sdo/193/{date}.fits')
sdo_euv_211 = Map(f'sdo/211/{date}.fits')
sdo_euv_304 = Map(f'sdo/304/{date}.fits')
sdo_euv_335 = Map(f'sdo/335/{date}.fits')
sdo_uv_1600 = Map(f'sdo/1600/{date}.fits')
sdo_uv_1700 = Map(f'sdo/1700/{date}.fits')
sdo_hmi_M720s = Map(f'sdo/hmi/{date}.fits').reproject_to(sdo_euv_171.wcs)
```

```python
fig = plt.figure(figsize=(10, 20))
ax = fig.add_subplot(5, 2, 1, projection=sdo_euv_94)
sdo_euv_94.plot(axes=ax, clip_interval=(1, 99.9)*u.percent)
ax = fig.add_subplot(5, 2, 2, projection=sdo_euv_131)
sdo_euv_131.plot(axes=ax, clip_interval=(1, 99.9)*u.percent)
ax = fig.add_subplot(5, 2, 3, projection=sdo_euv_171)
sdo_euv_171.plot(axes=ax, clip_interval=(1, 99.9)*u.percent)
ax = fig.add_subplot(5, 2, 4, projection=sdo_euv_193)
sdo_euv_193.plot(axes=ax, clip_interval=(1, 99.9)*u.percent)
ax = fig.add_subplot(5, 2, 5, projection=sdo_euv_211)
sdo_euv_211.plot(axes=ax, clip_interval=(1, 99.9)*u.percent)
ax = fig.add_subplot(5, 2, 6, projection=sdo_euv_304)
sdo_euv_304.plot(axes=ax, clip_interval=(1, 99.9)*u.percent)
ax = fig.add_subplot(5, 2, 7, projection=sdo_euv_335)
sdo_euv_335.plot(axes=ax, clip_interval=(1, 99.9)*u.percent)
ax = fig.add_subplot(5, 2, 8, projection=sdo_uv_1600)
sdo_uv_1600.plot(axes=ax, clip_interval=(1, 99.9)*u.percent)
ax = fig.add_subplot(5, 2, 9, projection=sdo_uv_1700)
sdo_uv_1700.plot(axes=ax, clip_interval=(1, 99.9)*u.percent)
ax = fig.add_subplot(5, 2, 10, projection=sdo_hmi_M720s)
sdo_hmi_M720s.plot(axes=ax, cmap='hmimag', norm=plt.Normalize(-1500, 1500))
plt.tight_layout()
plt.show()
```
